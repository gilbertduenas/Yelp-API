```python
# imports and froms
import json
import matplotlib.pyplot as plt
import pandas as pd
import requests

from yelp_key import api_key
from yelpapi import YelpAPI

# declare variables
all_results = pd.DataFrame()
yelp_api = YelpAPI(api_key)

# Import the census data into a Pandas DataFrame
zipcode_csv = pd.read_csv("uszips.csv")
zipcode_all = zipcode_csv[['zip']]

# get 500 random zip codes
zipcode_random = zipcode_all.sample(n=500)

print(f'Running loop.... This may take some time.')
for zipcode in zipcode_random['zip']:
    try: 
        search_results = yelp_api.search_query(location=zipcode, radius=4800, term='gyms')
        all_results = all_results.append(search_results['businesses'])
        print(zipcode)
    except:
        print(f'{zipcode}.... No businesses nearby')
```
